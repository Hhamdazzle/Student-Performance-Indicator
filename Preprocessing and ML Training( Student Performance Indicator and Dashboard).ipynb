# Count Missing Values in each column

```python
missing_data = df.isnull()
headers = df.columns
for column in headers:
  print(missing_data[column].value_counts())
  print(" ")
```

```python
df = df.fillna(0)
```

The dataset contained **no missing values**, so no rows were removed/dropped. Therefore, methods for handling missing values like replacing by mean, median, mode or frequency couldn't be performed.

##Handling non-numeric values in late_submission_counts column

```python
df['Late_submission_count (0-5)'] = (
    df['Late_submission_count (0-5)']
    .replace('5+', 5)
    .astype(float)
)
```

```python
df.info()
```



# **2.   Normalize Numeric Features**



Numeric features (assignment scores, test scores, attendance, study hours) normalized using MinMaxScaler to scale all values between 0 and 1.MinMax normalization was applied to academic and engagement-related numeric features to ensure all predictors lie on a comparable scale. Identifier and target variables were excluded. Standardization was not used as the dataset contained bounded values and MinMax scaling is more suitable for classification models and dashboard visualization.

```python
from sklearn.preprocessing import MinMaxScaler

numeric_cols = [
    'Assignment_Score 1 (out of 40)',
    'Assignment_Score 2(out of 40)',
    'Test_Result 1 (out of 10)',
    'Test_result 2 (out of 10)', 'Quiz_score (out of 7)', 'attendance_rate','midterm_score','weekly_self_study_hours',
    'class_participation','active_days_month'
]

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

```

```python
df.head()
```

```python
df[numeric_cols].dtypes
```

# **Creating Target**
Final scores alone do not show which students are at risk, so we derived performance_level from multiple indicators. This classification identifies at-risk students while ensuring the model predicts meaningful, actionable categories.

```python
df['performance_level'] = pd.cut(
    df[numeric_cols].mean(axis=1),
    bins=[0, 0.4, 0.7, 1.0],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['performance_level']
```



# **3.   Encoding Categorical Values with Scikit Learn**

# **OneHot Encoding with Scikit Learn**





```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output= False).set_output(transform = 'pandas')
ohetransform = ohe.fit_transform(df[['Course', 'Section']])
```

```python
df= pd.concat([df, ohetransform], axis= 1 ). drop(columns = ['Course', 'Section'])
```

# **Arranging Columns**
##The columns are arraned alphabetically after HotOne encoding, rearranging them for ease when dividing into predictor and target

```python
id_cols = ['student_id']
```

```python
target_col = ['performance_level']
```

```python
encoded_cols = ohetransform.columns.tolist()
```

```python
numeric_cols = [
    col for col in df.columns
    if col not in id_cols + target_col + encoded_cols
]
```

```python
df = df[id_cols + numeric_cols + encoded_cols + target_col]
```

```python
df.head(10)
```

```python
df.info()
```

# **Split Predictors And Target**

```python
X = df.iloc[ :, :-1]
X = X.drop(columns = ['student_id', 'final_score'])
```

```python
X
```

```python
y = df.iloc[:, -1]
```

```python
y
```




# **4. Split Data into Training And Testing**





```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 89)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```



# **B.   MACHINE LEARNING MODELS**






```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
results = []
```

# Applying Classification Algorithms





# 1.   Decision Trees






```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

results.append({
    'Model': 'Decision Tree',
    'Accuracy': accuracy_score(y_test, y_pred_dt),
    'Precision': precision_score(y_test, y_pred_dt, average='weighted'),
    'Recall': recall_score(y_test, y_pred_dt, average='weighted'),
    'F1-Score': f1_score(y_test, y_pred_dt, average='weighted')
})
```




# 2.   Random Forest



```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

results.append({
    'Model': 'Random Forest',
    'Accuracy': accuracy_score(y_test, y_pred_rf),
    'Precision': precision_score(y_test, y_pred_rf, average='weighted', zero_division=0),
    'Recall': recall_score(y_test, y_pred_rf, average='weighted', zero_division=0),
    'F1-Score': f1_score(y_test, y_pred_rf, average='weighted', zero_division=0)
})
```



# 3.   Logistic Regression




```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

results.append({
    'Model': 'Logistic Regression',
    'Accuracy': accuracy_score(y_test, y_pred_lr),
    'Precision': precision_score(y_test, y_pred_lr, average='weighted', zero_division= 0),
    'Recall': recall_score(y_test, y_pred_lr, average='weighted', zero_division=0),
    'F1-Score': f1_score(y_test, y_pred_lr, average='weighted', zero_division=0)
})

```



# 4.   Support Vector Machines(SVM)




```python
svm = SVC()
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

results.append({
    'Model': 'Support Vector Machine',
    'Accuracy': accuracy_score(y_test, y_pred_svm),
    'Precision': precision_score(y_test, y_pred_svm, average='weighted', zero_division=0),
    'Recall': recall_score(y_test, y_pred_svm, average='weighted', zero_division=0),
    'F1-Score': f1_score(y_test, y_pred_svm, average='weighted', zero_division=0)
})
```

# Comparison of Models

```python
results_df = pd.DataFrame(results)
results_df
```

# Best Performing Model

**Logistic Regression**

Multiple classification algorithms were applied, including Decision Tree, Random Forest, Logistic Regression, and Support Vector Machine. Model performance was evaluated using accuracy, precision, recall, and F1-score. Among all models, Logistic Regression achieved the highest accuracy and F1-score, indicating better generalization on the dataset. Hence, Logistic Regression was selected as the final model for deployment in the dashboard.
